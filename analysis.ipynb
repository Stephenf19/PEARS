# Demographics Analysis
This notebook aims to analyze the demographics data from Harry Chapin Food Bank Nutrition Education program. We will focus on various aspects such as age, gender, ethnicity, and location.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Loading
First, let's load the data into a Pandas DataFrame. For the purpose of this analysis, I'll assume that the data is available in a CSV file named `Program-Activities-Export (33) (1).csv`.

```python
# Load the data into a Pandas DataFrame
# Note: Replace 'demographics_data.csv' with the actual path to your CSV file
df = pd.read_csv('Program-Activities-Export (33) (1).csv')
```

## Age Distribution Analysis
We will start by analyzing the age distribution of participants in the programs.

```python
# Age Distribution Analysis
age_columns = ['participants_age_lt5', 'participants_age_5to17', 'participants_age_18to59', 'participants_age_60to75', 'participants_age_76plus']
age_distribution = df[age_columns].sum()
plt.figure(figsize=(12, 6))
sns.barplot(x=age_distribution.index, y=age_distribution.values, palette='viridis')
plt.title('Age Distribution of Participants')
plt.xlabel('Age Groups')
plt.ylabel('Number of Participants')
plt.show()
```

## Race Distribution Analysis
Next, we will analyze the distribution of participants based on their race.

```python
# Race Distribution Analysis
race_columns = ['participants_race_amerind', 'participants_race_asian', 'participants_race_black', 'participants_race_hawpac', 'participants_race_white']
race_distribution = df[race_columns].sum()
plt.figure(figsize=(12, 6))
sns.barplot(x=race_distribution.index, y=race_distribution.values, palette='mako')
plt.title('Race Distribution of Participants')
plt.xlabel('Race Groups')
plt.ylabel('Number of Participants')
plt.show()
```

## Ethnicity Distribution Analysis
We will now examine the distribution of participants based on their ethnicity.

```python
# Ethnicity Distribution Analysis
ethnicity_columns = ['participants_ethnicity_hispanic', 'participants_ethnicity_non_hispanic']
ethnicity_distribution = df[ethnicity_columns].sum()
plt.figure(figsize=(12, 6))
sns.barplot(x=ethnicity_distribution.index, y=ethnicity_distribution.values, palette='dark')
plt.title('Ethnicity Distribution of Participants')
plt.xlabel('Ethnicity Groups')
plt.ylabel('Number of Participants')
plt.show()
```

## Direct Education County Distribution Analysis
We will analyze the distribution of programs across different counties.

```python
# Analyzing the distribution of programs across different counties
county_distribution = df['site_county'].value_counts()

# Display the distribution
county_distribution
# Visualizing the distribution of programs across different counties
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.barh(county_distribution.index, county_distribution.values, color='blue')
plt.xlabel('Number of Direct Education Classes')
plt.ylabel('County')
plt.title('Direct Education by County')
plt.show()
```

## Participants per County Distribution Analysis
Finally, we will analyze the distribution of participants across different counties.

```python
# Analyzing the number of participants per county
participants_per_county = df.groupby('site_county')['participants_total'].sum()

# Display the number of participants per county
participants_per_county
# Importing necessary libraries for visualization
import matplotlib.pyplot as plt

# Creating a bar chart for the number of participants per county
plt.figure(figsize=(12, 6))
plt.barh(participants_per_county.index, participants_per_county.values, color='darkorange')
plt.xlabel('Number of Participants')
plt.ylabel('County')
plt.title('Number of Participants Per County')
plt.show()
```
